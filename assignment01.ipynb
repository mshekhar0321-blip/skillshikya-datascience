```python

books = [
    ("The Alchemist", "Fiction", 1988, 250),
    ("The Da Vinci Code", "Mystery", 2003, 300),
    ("A Brief History of Time", "Science", 1988, 150),
    ("The Theory of Everything", "Science", 2002, 100),
    ("Pride and Prejudice", "Fiction", 1813, 200),
    ("To Kill a Mockingbird", "Fiction", 1960, 180),
    ("The Catcher in the Rye", "Fiction", 1991, 220),
    ("Angels & Demons", "Mystery", 2000, 210),
    ("The Grand Design", "Science", 2010, 90),
    ("1984", "Fiction", 1949, 190)
]



```

```python
def filter_books(genre, year):
      filtered_books=[title for (title, g, y,t) in books if g==genre and y>=year] 
      return(filtered_books)
     
```

```python
print(filter_books("Fiction", 1980))
```

```python
#Control Flow with Nested Loops and Complex Logic
```

```python
while True:
    import random
    random_number=random.randint(1,100)
    attempts = 7
    print(f"Welcome! \n This game thinks of a number between 1 & 100.")
    print(f"you have {attempts} attempts.")
    for i in range (attempts):
        guess=int(input(f"attempts {i+1}: enter number:"))
        if guess == random_number:
            print("correct")
            break
        elif guess > random_number :
            print("low\n")
        else :
            print("high\n") 
    if guess != random_number:
       print(f"ran out of attempt. number was:{random_number}") 
    else:
       print(f"you are correct.number was:{random_number}")
   
    play_again=input("play again?(yes/no)")
    if play_again=="yes":
            print("good luck!")
            continue
    else:
            print("thank you")
            break
```
