```python
with open("finance_tools/tax.py", "w") as f:
    f.write("""
def calculate_tax(income):
    if income <= 500000:
        return 0
    elif income <= 1000000:
        return income * 0.1
    else:
        return income * 0.2
""")
```

```python
with open("finance_tools/loan.py", "w") as f:
    f.write("""
def calculate_emi(principal, rate, time):
    monthly_rate = rate / (12 * 100)
    months = time * 12
    emi = (principal * monthly_rate * (1 + monthly_rate) ** months) / \
          ((1 + monthly_rate) ** months - 1)
    return round(emi, 2)
""")
```

```python
from finance_tools.tax import calculate_tax
from finance_tools.loan import calculate_emi
```

```python
try:
    income=float(input("Enter your income:"))
    tax=calculate_tax(income)
    print("Tax to pay:",tax)

    principal=float(input("Enter your amount:"))
    rate=float(input("Enter interest rate%:"))
    time=float(input("Enter loan duration(year):"))
    emi=calculate_emi(principal, rate, time)
    print("Monthly emi:",emi)

except ValueError:
    print("Invalid input! enter numbers only")
    
```
