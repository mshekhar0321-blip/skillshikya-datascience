```python
class Vehical:
    def __init__(self, vehical_id:str, brand:str, rental_price:int):
        self.__vehical_id=vehical_id
        self.__brand=brand 
        self.__rental_price=rental_price

    def vehical_id(self):
        return self.__vehical_id

    def brand(self):
        return self.__brand

    def rental_price(self):
         return self.__rental_price

    def calculate_rental_price(self, days):
        return self.__rental_price * days

    def display_vehical_info(self):
        print("Vehical info")


class Car(Vehical):
    def __init__(self, vehical_id, brand, rental_price,number_of_doors:int):
            super().__init__(vehical_id, brand, rental_price)
            self.__number_of_doors=number_of_doors

    def number_of_doors(self):
            return self.__number_of_doors
         
    def display_vehical_info(self):
            print("Car info:")
            print(f"Vehical id= {self.vehical_id()}")
            print(f"Brand= {self.brand()}")
            print(f"Rental_price= {self.rental_price()}")
            print(f"Number_of_doors= {self.number_of_doors()}")
            print("total rental cost=", car.calculate_rental_price(5))

class Bike(Vehical):        
    def __init__(self, vehical_id, brand, rental_price,bike_type:str):
            super().__init__(vehical_id, brand, rental_price)
            self.__bike_type=bike_type

    def bike_type(self):
            return self.__bike_type

    def display_vehical_info(self):
            print("Bike info:")
            print(f"Vehical id= {self.vehical_id()}")
            print(f"Brand= {self.brand()}")
            print(f"Rental_price= {self.rental_price()}")
            print(f"Bike_type= {self.bike_type()}")
            print("total rental cost=", bike.calculate_rental_price(5))
    
            
    
        
```

```python
car= Car ("E200", "BMW", 5000,4)
car.display_vehical_info()
```
