```python
import math
import random 

class MathTutor:
    def __init__(self, questions=5):
       self.questions=questions
       self.correct=0
       self.incorrect=0

    def generate_question(self,*args):
        num1=random.randint(1,100)
        num2=random.randint(1,100)
        operator=random.choice(['+','-','/','*'])

        if operator=='+':
            question=  f"{num1}+{num2}"
            answer= num1 + num2
         
        elif operator=='-':
            question=  f"{num1}-{num2}"
            answer= num1 - num2
            
        elif operator=='/':
            question=  f"{num1}/{num2}"
            answer= round(num1 / num2,2)

        else :
            question=  f"{num1}*{num2}"
            answer= num1 * num2 

        
        return question,answer  


    def start_quiz(self):
        print("Welcme to MathTutor quiz")

        
        for i in range(self.questions):
           question,correct_answer=self.generate_question()
           print(f"Question {i+1}: {question}") 
           
           try:
              user_answer=float(input("Your answer:"))

              if round(user_answer,2)==correct_answer:
                  print("Correct")
                  self.correct+=1
              else: 
                  print(f"Incorrect.the correct answeris {correct_answer}")
                  self.incorrect+=1

           except ValueError:  
                 print("Invalid input!. Enter a number")
                 self.incorrect+=1
        
        self.show_score()


    def show_score(self):
       print("Quiz finished")
       print(f"score:{self.correct}/{self.questions}")
        
                   
        
                 
        

        
```

```python
tutor=MathTutor(questions=5)
tutor.start_quiz()
```

```python
class Product:
    def __init__(self,name:str,price:float):
        self.name=name
        self.price=price
```

```python
class User:
    def __init__(self,name:str,premium_user:bool):
        self.name=name
        self.premium_user=premium_user
```

```python
class Shopping_cart:
    def __init__(self,user:User):
        self.user=user
        self.cart=[]

    def add_product(self,product: Product):
        self.cart.append(product)
        print(f"{product.name} is added to cart")
        
    def remove_product(self,product: Product):
        if product in self.cart:
            self.cart.remove(product)
            print(f"{product.name} removed from cart")
        else:
            print("Product not found in cart.")
            
    def calculate_total_cost(self):
        total=sum(product.price for product in self.cart)

        if self.user.premium_user:
            total *=0.9
            
        return round(total,2)
```

```python
p1=Product("brush",40.0)
p2=Product("bottle",500.0)

user1=User("shekhar",True)

cart=Shopping_cart(user1)

cart.add_product(p1)
cart.add_product(p2)

cost=cart.calculate_total_cost()
print(f"total cost:{cost}")

```
